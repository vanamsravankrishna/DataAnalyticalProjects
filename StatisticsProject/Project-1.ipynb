### 1. Question: Please refer the table below to answer below questions:
| Planned to purchase Product A | Actually placed and order for Product A - Yes | Actually placed and order for Product A - No |     Marginal Total |
| ----------- | ------ | ------- | -------- |
| Yes | 400 | 100 | 500 |
| No | 200 | 1300 | 1500 |
| Marginal Total | 600 | 1400 | 2000 |


1. Refer to the above table and find the joint probability of the people who planned to purchase and actually placed an order.

2. Refer to the above table and find the joint probability of the people who planned to purchase and actually placed an order, given that people planned to purchase.

```python
# 1. Question
planned=500
placed_order=600
planned_and_placed_order=400
total_people=2000
p_pla=(planned/total_people) # probability of planned to purchase
p_pla_int_ord= (planned_and_placed_order/total_people) # intersection probability of planned and purchased
p_ord_giv_pla= (p_pla_int_ord/p_pla) # conditional probability(placed order given planned)
print("1. Joint probability of the people who planned to purchase and actually placed an order is %1.2f"%p_pla_int_ord)
print("2. Joint probability of the people who planned to purchase and actually placed an order, given that people planned to purchase is %1.2f"%p_ord_giv_pla)
```

### 2. Question: An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

A. Probability that none of the items are defective?

B. Probability that exactly one of the items is defective?

C. Probability that two or fewer of the items are defective?

D. Probability that three or more of the items are defective ?


```python
# 2. question
p_f=0.05
n=10
x=np.arange(11)
binom=stats.binom.pmf(x,n,p_f)
cubinom=stats.binom.cdf(x,n,p_f)
print("A. Probability that none of the items are defective is %1.4f"% binom[0])
print("B. Probability that exactly one of the items is defective is %1.4f"%binom[1])
print("C. Probability that two or fewer of the items are defective is %1.4f"%cubinom[2])
print("D. Probability that three or more of the items are defective is %1.4f"%(1-cubinom[2]))
```

### 3. Question: A car salesman sells on an average 3 cars per week.

A. Probability that in a given week he will sell some cars.

B. Probability that in a given week he will sell 2 or more but less than 5 cars.

C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

```python
# 3. Question
Lambda=3
k=np.arange(10)
poisson=stats.poisson.pmf(k,Lambda)
cupoisson=stats.poisson.cdf(k,Lambda)
print("A. Probability that in a given week he will sell some cars is %1.4f"%(1-poisson[0]))
print("B. Probability that in a given week he will sell 2 or more but less than 5 cars is %1.4f"%(cupoisson[4]-cupoisson[1]))
print("C. Plot of  poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.")
plt.figure(figsize=(10,5))
plt.plot(k,cupoisson,"o-")
plt.title("Poisson Distribution\n")
plt.xlabel("Number of cars sold per week")
plt.ylabel("Cumulative probability of cars sold per week")
plt.show()
```

### 5. Question: A group of 300 professionals sat for a competitive exam. The results show the information of marks obtained by them have a mean of 60 and a standard deviation of 12. The pattern of marks follows a normal distribution. Answer the following questions.

A. What is the percentage of students who score more than 80.

B. What is the percentage of students who score less than 50.

C. What should be the distinction mark if the highest 10% of students are to be awarded distinction?

```python
# 5. Question 
mu=60
std=12
n=300
x=np.arange(101)
z=(x-mu)/std
cunorm=stats.norm.cdf(z)
distinction_mark=round(stats.norm.isf(q =0.1, loc=60,scale=12),0)
print("A. The percentage of students who score more than 80 is %1.2f%s"%(((1-cunorm[80])*100),"%"))
print("B. The percentage of students who score less than 50 is %1.2f%s"%((cunorm[50]*100),"%"))
print("C. The distinction mark if the highest 10% of students are to be awarded distinction is {}".format(distinction_mark))
```

```python
# Handling column 11: Team Launch is a time period variable

sports_df["TeamLaunch"].replace(["_","to","~"], "-", inplace =True,regex=True)
```

```python
# defining a function so that we can use the team launch period variable in the further analysis
def variable_converter(x):
    if x.find("-")>=0:
        x =x.split("-")
    else:
        x= x.split()
    if len(x)==2:
        x[1]=x[0][0:2]+x[1]
    else:
        x.append(x[0])
    return(x)
```

```python
#separating the team start year and end year
sports_df["TeamLaunch"]=sports_df["TeamLaunch"].apply(variable_converter)
# creating new columns for start_year & start_2 Year
sports_df[['start_year','start_year_2']] = pd.DataFrame(sports_df.TeamLaunch.tolist(), index= sports_df.index)
#changing the team name by adding the year also so that we can have more info jst with the team name
sports_df["Team"]=sports_df["Team"]+"("+sports_df["start_year"]+")"
```

```python
# defining a function to get a summary of all missing in a given data.
# prints a string about no. of columns having missing values 
# returns a table with no. of missing values with its percentage of total no. of values for the columns which are 
#having missing values
def mis_val_table(df):
    mis_values_sum=df.isnull().sum()
    mis_val_percent=(df.isnull().sum()/len(df)).round(3)
    mis_val_table=pd.concat([mis_values_sum,mis_val_percent],axis=1)
    mis_val_table.rename(columns={0:"mis_values_sum",1:"mis_val_percent"},inplace=True)
    mis_val_table=mis_val_table[mis_val_table["mis_values_sum"]!=0]
    print("The given data has {} columns".format(df.shape[1]))
    print("Out of them, there are {} columns that have missing value".format(len(mis_val_table)))
    return(mis_val_table)
```

• Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?
          
          Write the null hypothesis and alternative hypothesis.
          Test for significance and conclusion

## Null hypothesis (Ho) : There is no significant difference in funding between the companies that are closed and companies that are still operating 

## Alternate hypothesis (Ha) : There is significant difference in funding between the companies that are closed and companies that are still operating 

#### Let's consider a significance level of 5%. i.e., alpha=0.05

```python
# by using t test for 2 independent samples
from scipy.stats import ttest_ind
```

```python
companies_still_operating=new_df[new_df["OperatingState"]=="Operating"]["funds_in_millions"]
companies_closed=new_df[new_df["OperatingState"]=="Closed"]["funds_in_millions"]
alpha=0.05
t_statistic, p_value= ttest_ind(companies_still_operating,companies_closed)
```

```python
if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha}(alpha) the difference in two groups is significant and we reject the Null hypothesis')
else:
    print(f'Since the p-value, {round(p_value,3)} >={alpha}(alpha) the difference in two groups is not significant and we fail to reject the Null hypothesis')
    
```

    Write the null hypothesis and alternative hypothesis.

## Null hyputhesis (Ho): The proportion of companies that are operating is the same in both categories - winners and contestants

## Alternative hypothesis (Ha): The proportion of companies that are operating is significantly different from each other, among the two categories

```python
new_df2=start_up_df.copy(deep=True) # Making a copy of the original data frame.
```

```python
# checking the frequency distribution of Result variable.
df1=new_df2["Result"].value_counts().reset_index()
df1.rename(columns={"count":"frequency"},inplace=True)
df1
```

```python
still_operating_results=new_df2[new_df2["OperatingState"]=="Operating"]["Result"]
```

```python
df2=still_operating_results.value_counts().reset_index();
df2.rename(columns={"count":"frequency"},inplace=True);
df2
```

```python
contestants_operating=(df2[df2.Result!="Winner"]["frequency"].sum())
winners_operating=(df2[df2.Result=="Winner"]["frequency"].sum())
contestants=(df1[df1.Result!="Winner"]["frequency"].sum())
winners=(df1[df1.Result=="Winner"]["frequency"].sum())
# considered only winner category as winner. and remaining all other categories as contestants 
#because remaining categories other than contestants are all not winners but they are also contastants
```

```python
print("contestants_operating=%d"%contestants_operating)
print("winners_operating=%d"%winners_operating)
print("contestants=%d"%contestants)
print("winners=%d"%winners)
```

```python
#Calculating percentage of winners that are still operating and total no. of winner 
P1=(winners_operating/winners)*100
print("percentage of winners that are still operating and total no. of winner is %1.2f"%P1)
```

```python
#calculating percentage of contestants that are still operating and total no. of contestants
P2=(contestants_operating/contestants)*100
print("percentage of contestants that are still operating and total no. of contestants is %1.2f"%P2)
```

```python
from statsmodels.stats.proportion import proportions_ztest

test_statistic, p_value = proportions_ztest([contestants_operating, winners_operating], [contestants, winners])

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

• Write and perform your hypothesis along with significance test comparing the funds raised by companies across NY, SF and EU events from 2013 onwards.

Each of the three above variables hold the amount of funds with respect to the Cities where the event was organised

## Null Hypothesis(Ho): Funds raised by companies across three cities Namely NewYork, Europe, Sanfransisco are the same 

## Alternative Hypothesis(Ha): Funds raised by companies across the events three cities are the different

```python
from scipy.stats import f_oneway


stat, p_value = f_oneway(NY_events, SF_events, EU_events)

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

```python
DF=pd.DataFrame()
df1=pd.DataFrame({"Event": "NY", "funds_in_millions":NY_events})
df2=pd.DataFrame({"Event": "EU", "funds_in_millions":EU_events})
df3=pd.DataFrame({"Event": "SF", "funds_in_millions":SF_events})
DF = pd.concat([DF, df1], ignore_index=True)
DF = pd.concat([DF, df2], ignore_index=True)
DF = pd.concat([DF, df3], ignore_index=True)
DF.sample(5)
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = "Event", y = "funds_in_millions", data = DF)
plt.title('Distribution of funding in million with Event ')
plt.show()
```



### Stating the null and alternative hypothesis:

* $H_0$: $\mu1$ = $\mu2$ = $\mu3$
* $H_A$: At least one $\mu$ differs 

Here $\mu1$, $\mu2$ and $\mu3$ are the mean company's funding with events took place at NY,EU,SF  respectively.

α = 0.05


```python
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
 
mod = ols('funds_in_millions ~ Event', data = DF).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

```python
P_value=aov_table["PR(>F)"].loc["Event"].round(3)
P_value # same as above f_oneway method
```

```python
if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```
